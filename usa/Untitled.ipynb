```python
import spacy
from spacy import displacy

# Load SpaCy's English language model
nlp = spacy.load("en_core_web_sm")

# Read the contents of the text file
with open("usa.txt", "r") as file:
    text = file.read()

# Process the text with SpaCy NER model
doc = nlp(text)

# Visualize the named entities
displacy.render(doc, style="ent", jupyter=True)
```

```python
# Extract named entities and their types
named_entities = [(entity.text, entity.label_) for entity in doc.ents]

# Print the named entities
for entity, entity_type in named_entities:
    print(f"Named Entity: {entity}, Entity Type: {entity_type}")
```
